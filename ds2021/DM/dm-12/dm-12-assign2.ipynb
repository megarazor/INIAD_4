### Curve fitting (vibration)

#### Import libraries  

```python
from numpy import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

#### Parameters  

```python
csv_in = 'vibration.csv'
%config InlineBackend.figure_formats = {'png', 'retina'}
```

#### Read data to be fitted from CSV file  

```python
df = pd.read_csv(csv_in, delimiter=',', skiprows=0, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

#### Scatter plot of data  

```python
plt.scatter(df['x'],df['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

#### Definition of function for fitting  

```python
def f(x, a, b, c):
    tot = a * np.exp(b * x) * np.sin(2 * np.pi * x + c)
    return tot
```

#### Execute fitting  

```python
popt, pcov = curve_fit(f, df['x'], df['y'])
```

```python
print(popt)
```

```python
y_reg = f(df['x'], *popt)  # *var: unpacking of list/tuple var
```

```python
plt.scatter(df['x'], df['y'], label='data')
lbl = 'f(x) = {0:.3f}e**{1:.3f}x * sin(2πx + {2:.3f})'.format(*popt)
plt.plot(df['x'], y_reg, color='red', label=lbl)
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
