```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
csv_in = 'cancer.csv'

# To show all rows and columns in the results 
pd.options.display.max_columns=999
pd.options.display.max_rows=999
```

```python
df = pd.read_csv(csv_in, delimiter=',', skiprows=14, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
print(df[df.Diag == 'B'].shape[0])
print(df[df.Diag == 'M'].shape[0])
```

```python
dfX = df.loc[:, 'ave_a':]
print(dfX.shape)
display(dfX.head())
```

```python
sc = StandardScaler()
X_std = sc.fit_transform(dfX)
```

```python
n_pca = 7
pca = PCA(n_components=n_pca)
X_pca = pca.fit_transform(X_std)
```

```python
print(X_pca.shape)
print(X_pca[:5])
```

```python
print(pca.explained_variance_ratio_)
print(np.cumsum(pca.explained_variance_ratio_))
```

Answer 2: 3 

```python
xx = range(1, n_pca+1)
plt.bar(xx, pca.explained_variance_ratio_)
plt.step(xx, np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.show()
```

```python
diag = df['Diag']
print(diag.value_counts())
```

```python
results = diag.unique()
print(results)
colors = ['red', 'green']
```

```python
pca_x = X_pca[:, 0]
pca_y = X_pca[:, 1]
for i in range(len(results)):
    cls = results[i]
    c = colors[i]
    plt.scatter(pca_x[diag==cls], pca_y[diag==cls],
                c=c, label=cls)
    print(cls, c)
plt.xlabel('PC_1')
plt.ylabel('PC_2')
plt.legend()
plt.show()
```

```python
def biplot(X_2d, coef_2d, coef_labels=None):
    r1 = 6
    r2 = 1.01
    coef_2dT = coef_2d.T
    if coef_labels is None:
        coef_labels = range(len(coef_2dT))
    for i, coef in enumerate(coef_2dT):
        plt.arrow(0, 0, coef[0]*r1, coef[1]*r1, color='r')    
        plt.text(coef[0]*r1*r2, coef[1]*r1*r2, coef_labels[i],
                 color='b', fontsize=11)
    plt.scatter(X_2d[:,0], X_2d[:,1])
    plt.xlabel('PC_1')
    plt.ylabel('PC_2')
    return None

biplot(X_pca[:, :2], pca.components_[:2], coef_labels=dfX.columns)
```
