```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
csv_in = 'ai-end2-3.csv'
```

```python
df = pd.read_csv(csv_in, delimiter=',', skiprows=0, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_data = df.loc[:, :'worst fractal dimension']
print(df_data.shape)
display(df_data.head())
```

```python
sc = StandardScaler()
X_std = sc.fit_transform(df_data)
```

```python
n_cls = 5
km = KMeans(n_clusters=n_cls, random_state=20)
cls = km.fit_predict(X_std)
```

```python
df_data['kmeans'] = cls
```

```python
print(df_data['kmeans'].value_counts())
```

```python
ct = pd.crosstab(df['diagnosis'], cls)
display(ct)
```

```python
df_ave = df_data.groupby('kmeans').mean()
display(df_ave)
```
