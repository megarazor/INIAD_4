```python
import pandas as pd
```

```python
csv1 = '1393_2017-utf8.csv'
csv2 = '2045_2017-utf8.csv'
```

```python
df1=pd.read_csv(csv1, delimiter=',', skiprows=9, header=0)
df1['Date']=pd.to_datetime(df1['Date'], format="%Y-%m-%d")
display(df1.head())

df2=pd.read_csv(csv2, delimiter=',', skiprows=9, header=0)
df2['Date']=pd.to_datetime(df2['Date'], format="%Y-%m-%d")
display(df2.head())

df12 = pd.merge(df1, df2, on='Date', how='inner')
display(df12.head())
```

```python
print(df12['High_x'].corr(df12['High_y']))
```
