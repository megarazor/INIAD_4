```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv('2045_2017-utf8.csv', delimiter=',', skiprows=9, header=0)
df1['Date']=pd.to_datetime(df1['Date'], format="%Y-%m-%d")
display(df1.head(3))

df2=pd.read_csv('1393_2017-utf8.csv', delimiter=',', skiprows=9, header=0)
df2['Date']=pd.to_datetime(df2['Date'], format="%Y-%m-%d")
display(df2.head(3))

df_merged = pd.merge(df1, df2, on='Date', how='inner'); display(df_merged.head(3))
```

```python
corr= df_merged['High_x'].corr(df_merged['High_y'])
print("Correlation coefficient: {:0.2f}\n".format(corr))
```
